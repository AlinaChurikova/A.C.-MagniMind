### Create a DataFrame whose index is the first 10 letters of the alphabet and contains two more columns with first 10 prime numbers and the first 10 fibonacci numbers.

```python
def Prime(n):  
    for i in range(2,n//2+1):  
        if(n%i==0):  
            return(0)  
    return(1)  
  
N=int(input("Enter N:"))  
i=2 
lst=[] 
while(1):  
    if(Prime(i)):  
        lst.append(i) 
        if(len(lst)==N): 
            break 
    i+=1 
print("First "+str(N)+" Prime numbers are:",end="") 
print(*lst)
df = pd.DataFrame(lst)
df
```

```python
df ["alphabet"] = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
df
```

```python
df.rename(columns={0: 'Prime Numbers'})
```

```python
new_df = df.set_index('alphabet')
```

```python
new_df 
```

```python
n=10
f=0                                         
sum=1                                        
print(f,sum,end=" ")
for x in range(2,n):
    next=f+sum                          
    print(next,end=" ")
    f=sum
    sum=next
```

```python
def fibonacci_nums(n):
    if n <= 0:
        return [0]
    sequence = [0, 1]
    while len(sequence) <= n:
        next_value = sequence[len(sequence) - 1] + sequence[len(sequence) - 2]
        sequence.append(next_value)
    return sequence
print("First 10 Fibonacci numbers:")
first_10_Fibonacci = fibonacci_nums(9)
print(first_10_Fibonacci)
```

```python
new_df ["first_10_Fibonacci"] = first_10_Fibonacci
new_df
```

### Merge these two DataFrames into a single DataFrame called census.

```python
import pandas as pd
census_division = pd.read_csv('census-divisions.csv')
state_population = pd.read_csv('state-populations.csv')

census = pd.merge(census_division, state_population)
```

```python
census
```

### Re-shape census so that one column contains all population measures, and another the year attributes.

```python
census.shape
```

```python
df = census.reset_index()
by_year = pd.melt(census, id_vars=['state',	'region',	'division'],  value_vars=['2010',	'2011',	'2012',	'2013',	'2014',	'2015',	'2016'])
by_year
```

```python
by_year2 = by_year.rename(columns={"variable": "year", "value": "population measures"})
```
